```python
from . import support
import time
```

```python
class Gun:
    attrName = "invenGun"
    def desc():
        support.writeConsole("It's a gun...")
        return False
    def shootSelf():
        support.writeConsole("I don't actually want to do that...")
        return False

class ConsumableGoods:
    attrName = "invenConsumableGoods"
    def desc():
        support.writeConsole("These are consumable goods, devoid of any interesting features. It's not obvious what can be done with them.")
        return False
    def eat():
        support.writeConsole("These consumable goods don't take kindly to people trying to eat them. Don't do that.")
        return False
    
class ArtisanalContract:
    attrName = "invenArtisanalContract"
    def desc():
        support.writeConsole("The missive's stamp of ivy vines, while broken, still evidences the detail of the seal.")
        return False
    def read():
        support.writeConsole("...", delay = 1)
        support.writeConsole("CHARGES INCURRED:")
        support.writeConsole("9X COMMON LABORER(S)")
        support.writeConsole("2X DESIGNER(S)")
        support.writeConsole("60X STICK(S) OF MOSS")
        support.writeConsole("DAMAGES:")
        support.writeConsole("2X COMMON LABORER(S)")
        time.sleep(1)
        support.writeConsole("...Very informative.")
        return False

class WhiteBook:
    attrName = "invenWhiteBook"
    def desc():
        support.writeConsole("An air of finality hangs over the book. Once you take this action, you know you won't be able to take it back - no matter how much you may want to...")
        support.writeConsole("...", delay = 1)
        support.writeConsole("Opening this book might not be the best idea.")
    def read():
        support.writeConsole("You fall, unwilling, unwanting to get up. Simply curling up for the rest of your days and dying on this floor doesn't seem so bad. All you have to do is let go of your worries...")
        time.sleep(5)
        support.writeConsole("BAD END")
        time.sleep(3)
        support.globalFlags["location"] = "End"
        return True
        
class BlackBook:
    attrName = "invenBlackBook"
    def desc():
        support.writeConsole("The book's heft is considerable and uncompromising, but not unkind.")
    def read():
        support.writeConsole("'where the need is window screens and sewer covers'")
        support.writeConsole("'where for the good of the very few and the suffering of a great many.'")
        time.sleep(1)
        support.writeConsole("...", delay = 1)
        support.writeConsole("The ruins will live.")
        support.globalFlags["location"] = "End"
        return True
    
class MistmossLeaflet:
    attrName = "invenMistmossLeaflet"
    def desc():
        support.writeConsole("It's a pristine leaflet, lacking the tears and dirt accompanying all its surroundings.")
    def read():
        support.writeConsole("...", delay = 1)
        support.writeConsole("Be aware of mistmoss!")
        support.writeConsole("Mistmoss is:")
        support.writeConsole("POISONOUS")
        support.writeConsole("COSTLY")
        support.writeConsole("ENVIRONMENTALLY HARMFUL")
        support.writeConsole("Join the movement!")
        time.sleep(1)
        support.writeConsole("...It ends there. Not particularly rousing.")
        support.writeConsole("..!", delay = 1)
        support.writeConsole("There's something scribbled messily on the back.")
        support.writeConsole("'where basically, everyone had a plant job'")
        support.writeConsole("'where preventable diseases rampant'")
        support.globalFlags["poemLines"][3] = 1
        return False
    
class About:
    attrName = "invenAbout"
    def desc():
        support.writeConsole("This game was made for a Literature assignment over two days as an interpretation of a poem! The poem used was [ruins library] by Myung Mi Kim.")
    def poem():
        fullPoem = """where the route of a ship bringing tax grain from the provinces is described
 
where perceived hindering—say, birds congregating on a runway

where the first request was for fertilizer and seeds
 
where the instruction—harness these to the benefit of your society
 
where the conscription continued
 
where boards of revenue
 
where basically, everyone had a plant job
 
where preventable diseases rampant
 
where the need is window screens and sewer covers
 
where for the good of the very few and the suffering of a great many"""
        support.writeConsole(fullPoem, delay = .07)
        return False
```
